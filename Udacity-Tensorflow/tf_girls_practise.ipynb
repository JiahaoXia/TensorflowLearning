**Tensorflow中的4个重要类型**


---
*   Variable 变量，类型为class
*   Tensor 多维矩阵，带有多种方法，类型为class
*   Graph 计算图谱
*   Session 用于运行一个计算图谱

**Tensorflow中的3个重要函数**


---


*   tf.Variable
*   tf.constant 
*   tf.placeholder return:一个尚未存在的tensor







```
import tensorflow as tf
```

**Basic Operation**

```
v1 = tf.Variable(10)
v2 = tf.Variable(5)
addv = v1 + v2
print(addv)
print(type(v1))
print(type(addv))

c1 = tf.constant(10)
c2 = tf.constant(5)
addc = c1 + c2

# session is a runtime
sess = tf.Session()
# variable -> initialize -> given 
tf.initialize_all_variables().run(session=sess)
print('variable need initialization')
print('v1 + v2 = ', addv.eval(session=sess))
print('v1 + v2 = ',sess.run(addv))
print('c1 + c2 = ',sess.run(addc))
print(addc)
```

```
# symbolic programming
graph = tf.Graph()
with graph.as_default():
  value1 = tf.constant([1,2])
  value2 = tf.constant([3,4])
  mul = value1 * value2

with tf.Session(graph=graph) as mySess:
  tf.initialize_all_variables().run()
  print('value1 * value2 = ', mySess.run(mul))
  print('/n')
  print('value1 * value2 = ',mul.eval())
```

**Placeholder**
